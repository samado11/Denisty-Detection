```python
def save_density_map(density_map, name):
    plt.figure(dpi=250)
    plt.axis('off')
    plt.margins(0, 0)
    plt.imshow(density_map, cmap=CM.jet)
    plt.savefig(name, dpi=250, bbox_inches='tight', pad_inches=0)
```
